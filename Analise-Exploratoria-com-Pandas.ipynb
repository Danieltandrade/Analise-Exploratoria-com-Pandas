```python
# Importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="darkgrid")  # Configurando o estilo do seaborn
```

```python
# Criando nosso DataFrame
df = pd.read_excel('AdventureWorks.xlsx')
```

```python
# Visualizando as 5 primeiras linhas
df.head()
```

```python
# Quantidade de linhas e colunas
df.shape
```

```python
# Verificando os tipos de dados
df.dtypes
```

```python
# Qual a receita total?
df['Valor Venda'].sum()
```

```python
# Qual o custo total?
df['Custo'] = df['Custo Unitário'].mul(df['Quantidade']) # Criando coluna de custo
```

```python
df.head(2)
```

```python
# Qual o custo total?
round(df['Custo'].sum(), 2)
```

```python
# Criando coluna de Lucro.
df['Lucro'] = df['Valor Venda'] - df['Custo']
```

```python
df.head(2)
```

```python
# Lucro Total.
round(df['Lucro'].sum(), 2)
```

```python
# Criando coluna com total de dias para enviar o produto.
df['Tempo_envio'] = df['Data Envio'] - df['Data Venda']
```

```python
df.head(2)
```

**Média de tempo de envio para cada marca.**

```python
# Extraindo apenas os dias.
df['Tempo_envio'] = (df['Data Envio'] - df['Data Venda']).dt.days
```

```python
df.head(2)
```

```python
# Verificando o tipo de coluna envio.
df['Tempo_envio'].dtype
```

```python
# Média do tempo de envio por marca.
df.groupby('Marca')['Tempo_envio'].mean()
```

**Valores faltantes.**

```python
# Verificando se temos dados faltantes.
df.isnull().sum()
```

**Lucro por ano e marca?**

```python
# Agrupando por ano e marca.
df.groupby([df['Data Venda'].dt.year, 'Marca'])['Lucro'].sum()
```

```python
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
# Resetando o index.
lucro_ano = df.groupby([df['Data Venda'].dt.year, 'Marca'])['Lucro'].sum().reset_index()
lucro_ano
```

```python
# Qual o total de produtos vendidos.
df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=False)
```

```python
# Gráfico total de produtos vendidos.
df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=True).plot.barh(title='Total de Produtos Vendidos')
plt.xlabel('Total')
plt.ylabel('Produto');
```

```python
# Lucro por ano.
df.groupby(df['Data Venda'].dt.year)['Lucro'].sum().plot.bar(title='Lucro por Ano')
plt.xlabel('Ano')
plt.ylabel('Receita');
```

```python
df.groupby(df['Data Venda'].dt.year)['Lucro'].sum()
```

```python
# Selecionando apenas as vendas de 2009
df_2009 = df[df['Data Venda'].dt.year == 2009]
```

```python
df_2009.head()
```

```python
# Grafico de linhas: Lucro x Mês
df_2009.groupby(df_2009['Data Venda'].dt.month)['Lucro'].sum().plot(title='Lucro x Mês')
plt.xlabel('Mês')
plt.ylabel('Lucro');
```

```python
# Grafico de barras: Lucro x Marca
df_2009.groupby('Marca')['Lucro'].sum().plot.bar(title='Lucro x Marca')
plt.xlabel('Marca')
plt.ylabel('Lucro')
plt.xticks(rotation='horizontal');
```

```python
# Grafico de barras: Lucro x Classe
df_2009.groupby('Classe')['Lucro'].sum().plot.bar(title='Lucro x Classe')
plt.xlabel('Classe')
plt.ylabel('Lucro')
plt.xticks(rotation='horizontal');
```
